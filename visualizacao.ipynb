Este notebook executa a leitura de uma instância `.dat`, calcula métricas do grafo e exibe resultados de forma tabular e gráfica.

Suba os arquivos `grafo.py`, `parser.py`, `estatisticas.py` e o `.dat` desejado no Colab para rodar as células abaixo.

As métricas são: quantidade de vértices, arcos/arestas, densidade, grau, intermediação, diâmetro, etc.

```python
import pandas as pd
import matplotlib.pyplot as plt
from parser import ler_arquivo_dat
from estatisticas import calcular_estatisticas

# Substitua pelo nome do arquivo .dat que você enviou, se for diferente
arquivo = 'mgval_0.50_10D.dat'

grafo = ler_arquivo_dat(arquivo)
stats = calcular_estatisticas(grafo)

# Separar intermediação do resto para visualização
intermediacao = stats.pop('Intermediacao')

# Mostrar as estatísticas principais como tabela
df = pd.DataFrame([stats])
df
```

```python
plt.figure(figsize=(10, 4))
plt.bar(intermediacao.keys(), intermediacao.values(), color='mediumseagreen')
plt.title('Centralidade por Intermediação')
plt.xlabel('Vértices')
plt.ylabel('Intermediação')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
